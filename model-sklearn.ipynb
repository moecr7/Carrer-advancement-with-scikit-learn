```python
import numpy as np 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv("Carrer-Advancement.csv")#, delimiter=","
df.iloc[:,9:]
```

```python
df.columns
```

```python
df['learning_style'].value_counts()
```

```python
df.fillna(0, inplace=True)
```

```python
df.iloc[:,7:15]
```

```python
from sklearn.model_selection import train_test_split
```

```python
print(df['long_term_encoded'])
print(df['long_term_goal'])
```

```python
X = df[['age','gender_encoded', 'learning_style_encoded', 'current_role_encoded',
       'current_industry_encoded', 'previous_roles_encoded',
       'desired_role_encoded', 'preferred_industry_encoded',
       'short_term_encoded', 'long_term_encoded', 'key_skills_encoded',
       'recommended_skills_encoded', 'certifications_encoded',]]
```

```python
y = df[['recommended_skills_encoded']]
```

```python
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.2)
```

```python
des = DecisionTreeClassifier(criterion="log_loss", max_depth = 6)
```

```python
des.fit(X_trainset,y_trainset)
```

```python
predTree = des.predict(X_testset)
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, predTree))
```
